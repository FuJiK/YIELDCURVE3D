```python
import pandas as pd
import plotly.express as px
from fredapi import Fred
from credential.api_key import FRED_API_KEY  # api_key.pyからAPIキーをインポート

fred = Fred(api_key=FRED_API_KEY)
```

```python
start_date = "2020-01-01"  # 開始日を指定
end_date = "2023-12-01"  # 終了日を指定
```

```python
# 各要素のCPIシリーズIDとカテゴリ名
cpi_series = {
    "CPIAUCSL": "全アイテム（アメリカ全体）",
    "CPIFABSL": "食品と飲料",
    "CPIUFDSL": "食品",
    # "CPIDFESL": "食品（家庭外）",
    "CPIHOSSL": "住宅",
    "CPIAPPSL": "着衣",
    "CPITRNSL": "輸送",
    "CPIMEDSL": "医療",
    # "CUSR0000SAG": "他の商品とサービス"
}
```

```python
# 各シリーズIDのデータを取得し、データフレームに統合
df = pd.DataFrame()
for series_id, category in cpi_series.items():
    series_data = fred.get_series(series_id, start_date, end_date)
    temp_df = pd.DataFrame(series_data, columns=[category])
    temp_df = temp_df.reset_index().rename(columns={"index": "年月"})
    
    if df.empty:
        df = temp_df
    else:
        df = df.merge(temp_df, on="年月")
```

```python
# 時系列データの可視化
fig = px.line(df, x="年月", y=df.columns[1:], title="各要素の消費者物価指数の遷移")
fig.show()
```
