```python
import quandl
import plotly.graph_objects as go
from datetime import datetime, timedelta
from credential.api_key import QUANDL_API_KEY
```

```python
# Quandl APIキーを設定
quandl.ApiConfig.api_key = QUANDL_API_KEY # ここにAPIキーを入力
```

```python
# 過去一年分の日付を設定
end_date = datetime.now()
start_date = end_date - timedelta(days=365)
```

```python
# QuandlからCOTレポートのNQ100先物情報を取得
cot_data = quandl.get('CFTC/209742_FO_ALL', start_date=start_date, end_date=end_date)
```

```python
# Producer/Merchant/Processor/User LongsとSwap Dealer LongsとMoney Manager Longsの和を計算し、新しいカラムを追加
cot_data['Long_netposition'] = cot_data['Dealer Longs'] + cot_data['Asset Manager Longs'] + cot_data['Leveraged Funds Longs']

# Producer/Merchant/Processor/User ShortsとSwap Dealer ShortsとMoney Manager Shortsの和を計算し、新しいカラムを追加
cot_data['Short_netposition'] = cot_data['Dealer Shorts'] + cot_data['Asset Manager Shorts'] + cot_data['Leveraged Funds Shorts']
```

```python
cot_data
```

```python
# データをプロット
fig = go.Figure()

fig.add_trace(go.Scatter(x=cot_data.index, y=cot_data['Open Interest'],
                    mode='lines', name='Open Interest'))
fig.add_trace(go.Scatter(x=cot_data.index, y=cot_data['Long_netposition'],
                    mode='lines', name='Long_netposition'))
fig.add_trace(go.Scatter(x=cot_data.index, y=cot_data['Short_netposition'],
                    mode='lines', name='Short_netposition'))
fig.add_trace(go.Scatter(x=cot_data.index, y=cot_data['Dealer Longs'],
                    mode='lines', name='Dealer Longs'))
fig.add_trace(go.Scatter(x=cot_data.index, y=cot_data['Dealer Shorts'],
                    mode='lines', name='Dealer Shorts'))
fig.add_trace(go.Scatter(x=cot_data.index, y=cot_data['Asset Manager Longs'],
                    mode='lines', name='Asset Manager Longs'))
fig.add_trace(go.Scatter(x=cot_data.index, y=cot_data['Asset Manager Shorts'],
                    mode='lines', name='Asset Manager Shorts'))
fig.add_trace(go.Scatter(x=cot_data.index, y=cot_data['Leveraged Funds Longs'],
                    mode='lines', name='Leveraged Funds Longs'))
fig.add_trace(go.Scatter(x=cot_data.index, y=cot_data['Leveraged Funds Shorts'],
                    mode='lines', name='Leveraged Funds Shorts'))

fig.update_layout(title='COT Report - NQ100 Futures',
                  xaxis_title='Date',
                  yaxis_title='Positions',
                  legend_title='Positions')

fig.show()
```
